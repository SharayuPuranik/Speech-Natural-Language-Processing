```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

name_variations_df = pd.read_csv('/kaggle/input/textsearchtask2/name_variations (1).csv')
base_names_df = pd.read_csv('/kaggle/input/textsearchtask2/base_names (1).csv')
```

```python
name_variations_df['Normalized_Variation'] = name_variations_df['Variation'].str.lower().str.strip()
base_names_df['Normalized_Base_Name'] = base_names_df['Base_Name'].str.lower().str.strip()

def fuzzy_match(name, base_names):
    best_match = process.extractOne(name, base_names, scorer=fuzz.token_sort_ratio)
    if best_match is None:
        return ("No Match", 0)
    return best_match

name_variations_df['Fuzzy_Match'] = name_variations_df['Normalized_Variation'].apply(
    lambda x: fuzzy_match(x, base_names_df['Normalized_Base_Name'])
)

name_variations_df['Best_Match'] = name_variations_df['Fuzzy_Match'].apply(lambda x: x[0])
name_variations_df['Match_Score'] = name_variations_df['Fuzzy_Match'].apply(lambda x: x[1])

name_variations_df.drop(columns=['Fuzzy_Match'], inplace=True)

name_variations_df['Expected_Match_Correct'] = name_variations_df['Matches_With_Base_Name'].str.lower().str.strip() == name_variations_df['Best_Match']

name_variations_df.head(20)
```
